Getting back to my main objective in this project: comparing the influence of each marketing campaign on the customers.A problem may arise as customers can respond simultaneously to more than 1 campaign.To solve this, I have duplicated records when customers engage with several campaigns at the same time.

```python
import warnings

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    dfa1=df[df['AcceptedCmp1']>0]
    dfa1.loc[:,'AcceptedCmp']=1
    dfa2 = df[df['AcceptedCmp2']>0]
    dfa2.loc[:,'AcceptedCmp'] = 2
    dfa3 = df[df['AcceptedCmp3']>0]
    dfa3.loc[:,'AcceptedCmp'] = 3
    dfa4 = df[df['AcceptedCmp4']>0]
    dfa4.loc[:,'AcceptedCmp'] = 4
    dfa5 = df[df['AcceptedCmp5']>0]
    dfa5.loc[:,'AcceptedCmp'] = 5
```

```python
#Combining in 1 dataset

dfa = pd.concat([dfa1, dfa2, dfa3, dfa4, dfa5])
dfa
```

Now, Let's group the campaigns and analyze:

```python
df_p=pd.pivot_table(dfa,values='ID',index=['AcceptedCmp','Education'],
                   columns=['Marital_Status'],aggfunc="count",fill_value=0, margins=True,margins_name="Total")
df_p
```

Let's calculate the percentage of sales per campaign:

```python
pd.options.display.float_format = '{:.1f}%'.format
dfa['AcceptedCmp'].value_counts(normalize=True).sort_index()*100
```

- The third,forth and fifth campaigns have the most influence on the sales.
- The second campaign is not effective.
<br>In the following code, I am conducting a benchmark and comparing how different are the results of the 1st campaign from the rest.

```python
d=df_p.reset_index()
#Deleting the total column 
d=d[d.columns[:-1]]
#creating an index by the field Education
d=df_p.reset_index()
d=d.set_index('Education')
d
```

```python
#Calculating the difference between the campaigns 
df_dif=pd.DataFrame()
for i in range(2,6):
    df_dif= df_dif.append(d[d['AcceptedCmp']==i]-d[d['AcceptedCmp']==1])
df_dif['AcceptedCmp']= df_dif['AcceptedCmp']+ 1
df_dif.groupby(['AcceptedCmp','Education'],).sum()
```

The second campaign contains mostly negative values.This proves its ineffienciency compared to the first campaign.

### Marketing campaigns impact

In this section, I am going to visualize the pivot tables obtained in the previous section and display the results of each marketing campaign

```python
# Filling empty values of "Total" column with 0
df_pie = df_p.reset_index()
df_pie.Total = df_pie.Total.fillna(0)
```

```python
# Building pie charts to visualize the distribution of people based on education who bought the product
# for each marketing campaign
def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n{:d}".format(pct, absolute)

# Loop through each marketing campaign (AcceptedCmp) from 1 to 5
for e in range(1, 6):
    plt.figure(figsize=(12,5))
    dt = df_pie[df_pie['AcceptedCmp']==e]
     
    exp = np.zeros(len(dt))
    exp[np.argmax(dt.Total)] = 0.1

    plt.pie(dt.Total, shadow = True,   explode=exp, 
            autopct=lambda pct: func(pct, dt.Total),
            textprops=dict(color="w"))

    plt.title('AcceptedCmp ' + str(e))
    plt.legend(dt.Education, title="Education",
          loc="center right", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.show()
```
